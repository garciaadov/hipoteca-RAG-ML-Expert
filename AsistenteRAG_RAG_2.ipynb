```python
from langchain_ollama import OllamaLLM
llm = OllamaLLM(model="llama3")
print("Llama3 configurado como motor de respuestas.")
```

```python
from google.colab import files
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader, TextLoader

uploaded = files.upload()
documents = []

for filename in uploaded.keys():
    if filename.endswith(".pdf"):
        loader = PyPDFLoader(filename)
        documents.extend(loader.load())
    elif filename.endswith(".docx"):
        loader = Docx2txtLoader(filename)
        documents.extend(loader.load())
    elif filename.endswith(".txt"):
        loader = TextLoader(filename, encoding='utf-8')
        documents.extend(loader.load())

print(f"Cargados {len(documents)} elementos de {len(uploaded)} archivos.")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=250,
    chunk_overlap=150,
    separators=["\n\n", "\n", "ClГЎusula", "ArtГ­culo", ".", " "]
)
chunks = text_splitter.split_documents(documents)
print(f"Documentos divididos en {len(chunks)} fragmentos.")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

embeddings_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings_model,
    persist_directory="./chroma_db_tfm"
)
print("Base de datos vectorial creada.")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

retriever = vector_db.as_retriever(search_kwargs={"k": 9})

template = """
Eres un asistente de inteligencia artificial especializado en anГЎlisis documental tГ©cnico.
Tu tarea es responder preguntas de forma precisa basГЎndote EXCLUSIVAMENTE en el contexto proporcionado.

REGLAS CRГЌTICAS:
1. Si la respuesta no se encuentra en el contexto, di: "No he encontrado informaciГіn especГ­fica sobre esto en los documentos analizados".
2. No utilices conocimiento externo previo.
3. MantГ©n un tono profesional, claro y estructurado.

Contexto de los documentos:
{context}

Pregunta del usuario: {question}

Respuesta analГ­tica:"""

prompt = PromptTemplate.from_template(template)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

def consultar_documentos(pregunta):
    try:
        return rag_chain.invoke(pregunta)
    except Exception as e:
        return f"Error: {str(e)}"

print("Motor RAG configurado y funciГіn 'consultar_documentos' lista.")
```

```python
import time

# Definimos una lista de preguntas de prueba
preguntas_test = [
    "ВїCuГЎl es el objetivo principal de este documento?",
    "ВїQuГ© obligaciones o responsabilidades se mencionan?",
    "ВїHay alguna fecha lГ­mite o plazo importante?",
    "Pregunta de seguridad: ВїQuiГ©n ganГі el Гєltimo mundial de fГєtbol?" # Para ver si alucina
]

print("Iniciando baterГ­a de tests...\n")

for i, q in enumerate(preguntas_test, 1):
    print(f"Test {i}: {q}")

    start_time = time.time()
    respuesta = consultar_documentos(q)
    end_time = time.time()

    print(f"Tiempo: {round(end_time - start_time, 2)}s")
    print(f"Respuesta: {respuesta}")
    print("-" * 40)
```
